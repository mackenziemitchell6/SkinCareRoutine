```python
full=pd.concat([reddf,sensitivedf,drydf,oilydf,ncdf,darkcdf,agedf,acnedf])
full.drop_duplicates(subset='prodName',inplace=True)
```

```python
antiagee=[]
for f in full.prodName:
    if f in age:
        antiagee.append(1)
    else:
        antiagee.append(0)
circle=[]
for f in full.prodName:
    if f in darkcircles:
        circle.append(1)
    else:
        circle.append(0)
acn=[]
for f in full.prodName:
    if f in acne:
        acn.append(1)
    else:
        acn.append(0)
dr=[]
for f in full.prodName:
    if f in dryness:
        dr.append(1)
    else:
        dr.append(0)
re=[]
for f in full.prodName:
    if f in redness:
        re.append(1)
    else:
        re.append(0)
senss=[]
for f in full.prodName:
    if f in sensi:
        senss.append(1)
    else:
        senss.append(0)
oi=[]
for f in full.prodName:
    if f in oily:
        oi.append(1)
    else:
        oi.append(0)
nor=[]
for f in full.prodName:
    if f in norm:
        nor.append(1)
    else:
        nor.append(0)
```

```python
ty=[]
for f in full.prodName:
    if f in norm:
        ty.append(1)
    elif f in oily:
        ty.append(2)
    elif f in sensi:
        ty.append(3)
    elif f in redness:
        ty.append(4)
    elif f in dryness:
        ty.append(5)
    elif f in acne:
        ty.append(6)
    elif f in darkcircles:
        ty.append(7)
    elif f in age:
        ty.append(8)
```

```python
full['age']=antiagee
full['darkcircles']=circle
full['acne']=acn
full['dry']=dr
full['redness']=re
full['sensitive']=senss
full['oil']=oi
full['normal']=nor
full['treats']=ty
```
