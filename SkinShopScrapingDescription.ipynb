```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from bs4 import BeautifulSoup
import pickle
```

```python
with open('fulldataframe.pickle', 'rb') as f:
    # The protocol version used is detected automatically, so we do not
    # have to specify it.
    full = pickle.load(f)
```

```python
descriptiondict=[]
for i in full.url:
    response=requests.get('https://www.skinstore.com/slip-silk-pillowcase-queen-white/{}.html'.format(i))
    soup=BeautifulSoup(response.content,'html.parser')
    description=soup.findAll('div',{'class':'productDescription_synopsisContent'})
    describe=''
    for d in description:
        describe=d.text.replace('\n','')
    descriptiondict.append({'url':i,'description':describe})
descriptiondf=pd.DataFrame(descriptiondict)
```

```python
descriptiondf.description=str(descriptiondf.describe)
```
